```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
X
```

```python
y
```

```python
import sklearn.svm as svm

svm_clf =svm.SVC(kernel = 'linear')
```

```python
from sklearn.model_selection import cross_val_score, cross_validate

scores = cross_val_score(svm_clf, X, y, cv = 5)
scores
```

```python
import pandas as pd

pd.DataFrame(cross_validate(svm_clf, X, y, cv =5))
```

```python
print('교차검증 평균: ', scores.mean())
```

```python
svm_clf1 =svm.SVC(kernel = 'rbf')

# 교차검증

scores1 = cross_val_score(svm_clf1, X, y, cv = 5)
scores1

pd.DataFrame(cross_validate(svm_clf1, X, y, cv =5))

print('교차검증 평균: ', scores1.mean())
```

```python
from sklearn.preprocessing import StandardScaler

# StandarScaler 적용

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
print(X[0:4])
```

```python
print(X_scaled[0:4])
```

```python
svm_clf_standard =svm.SVC(kernel = 'linear')

scores2 = cross_val_score(svm_clf_standard, X_scaled, y, cv = 5)

print('교차검증 평균: ', scores2.mean())
```

```python
#비선형 + 표준정규화

svm_clf_scaled2 =svm.SVC(kernel = 'rbf')

# 교차검증

scores3 = cross_val_score(svm_clf_scaled2, X_scaled, y, cv = 5)
scores3

print('교차검증 평균: ', scores3.mean())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)

X_minmax = scaler.transform(X)
```

```python
svm_clf_minmax =svm.SVC(kernel = 'linear')

scores4 = cross_val_score(svm_clf_minmax, X_minmax, y, cv = 5)

print('교차검증 평균: ', scores4.mean())
```

```python
svm_clf_minmax2 =svm.SVC(kernel = 'rbf')

# 교차검증

scores5 = cross_val_score(svm_clf_minmax2, X_minmax, y, cv = 5)

print('교차검증 평균: ', scores5.mean())
```
